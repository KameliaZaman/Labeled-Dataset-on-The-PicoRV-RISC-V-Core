# Extract output value (e.g., x12) and compute intermediate model

```python
# Define leakage models
def HW(x):
    return bin(x).count('1')

def HD(x, y):
    return bin(x ^ y).count('1')

# Convert hex strings to integers
x12_int = input_data["x12"].apply(lambda x: int(x, 16))
x15_int = input_data["x15"].apply(lambda x: int(x, 16))

# ------------------
# Hamming Weight (HW)
# ------------------
reg_sum_hw = x12_int + x15_int
intermediates_list_hw = reg_sum_hw.apply(HW).to_numpy()[:100]

# ------------------
# Hamming Distance (HD)
# ------------------
reg_sum_hd = reg_sum_hw.copy()
reg_sum_shifted = reg_sum_hd.shift(1).fillna(0).astype(int)  # previous value
reg_sum_hd = reg_sum_hd.astype(int)                          # current value

intermediates_list_hd = np.array([
    HD(curr, prev) for curr, prev in zip(reg_sum_hd, reg_sum_shifted)
])[:100]

# Check shapes
print("HW shape:", intermediates_list_hw.shape)
print("HD shape:", intermediates_list_hd.shape)
```

# Add Gaussian Noise

```python
def add_noise(traces, snr_db):
    """
    Add Gaussian noise to traces based on the given SNR in dB.
    SNR (Signal-to-Noise Ratio) = signal_power / noise_power
    """
    signal_power = np.mean(traces ** 2)
    snr_linear = 10 ** (snr_db / 10)
    noise_power = signal_power / snr_linear
    noise = np.random.normal(0, np.sqrt(noise_power), traces.shape)
    return traces + noise

# Add noise (e.g., SNR = 5 dB; adjust for more or less noise)
trace_data_HW = [add_noise(traces, snr_db=20) for traces in trace_data_HW]

trace_data_HD = [add_noise(traces, snr_db=20) for traces in trace_data_HD]
```

# Normalize traces

```python
trace_data_HW = [
    (td - td.mean(axis=0)) / td.std(axis=0) for td in trace_data_HW
]

trace_data_HD = [
    (td - td.mean(axis=0)) / td.std(axis=0) for td in trace_data_HD
]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# 1. Prepare all traces (8 traces × 100 samples)
all_traces = np.array(trace_data_HW)  # shape: (8, 100)

# 2. Prepare intermediates (already verified correct)
intermediates = np.array(intermediates_list_hw[:8])  # 8 values for 8 traces

# 3. Time-sample analysis with statistical significance
significant_points = []
correlation_series = []

for sample_idx in range(100):
    # Get all traces' values at this time sample
    sample_values = all_traces[:, sample_idx]
    
    # Compute correlation and p-value
    r, p_value = pearsonr(sample_values, intermediates)
    correlation_series.append(r)
    
    if p_value < 0.05:  # 95% confidence level
        significant_points.append(sample_idx)

# 4. Visualization
plt.figure(figsize=(12, 5))
plt.plot(correlation_series, label='Correlation')
plt.scatter(significant_points, 
           [correlation_series[i] for i in significant_points],
           color='red', label='Significant (p < 0.05)')
plt.axhline(0, color='black', linestyle=':')
plt.title("Correlation Analysis Over Time")
plt.xlabel("Time Sample")
plt.ylabel("Pearson r")
plt.legend()
plt.show()

# 5. Multi-trace analysis
trace_correlations = []
for trace_idx in range(8):
    r, _ = pearsonr(all_traces[trace_idx], intermediates)
    trace_correlations.append(r)
    print(f"Trace {trace_idx} correlation: {r:.4f}")

print(f"\nAverage correlation: {np.mean(trace_correlations):.4f}")
print(f"Maximum correlation: {np.max(trace_correlations):.4f}")
```

```python
# All traces: shape (8, 100)
# Intermediates: shape (8,)

# Correlation at each time point across traces
time_sample_correlations = []
for t in range(100):  # For each time sample
    r, _ = pearsonr(all_traces[:, t], intermediates)
    time_sample_correlations.append(r)

plt.plot(time_sample_correlations)
plt.title("Correlation at Each Time Sample")
plt.xlabel("Time Sample")
plt.ylabel("Pearson r")
plt.show()
# Enhanced analysis with statistics
significant_points = []
p_values = []

for t in range(100):
    r, p = pearsonr(all_traces[:, t], intermediates)
    time_sample_correlations.append(r)
    p_values.append(p)
    if p < 0.05:
        significant_points.append(t)

# Visualization
plt.figure(figsize=(12,5))
plt.plot(time_sample_correlations, label='Correlation')
plt.scatter(significant_points, 
           [time_sample_correlations[i] for i in significant_points],
           color='red', label='p < 0.05')
plt.axhline(0, color='k', linestyle='--')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.signal import savgol_filter, medfilt

# 1. Normalize each trace individually
normalized_traces = (traces - np.mean(traces, axis=1, keepdims=True)) / np.std(traces, axis=1, keepdims=True)

# 2. Apply Savitzky-Golay filter (preserves peaks)
filtered_traces = np.array([savgol_filter(t, window_length=5, polyorder=2) for t in normalized_traces])

# 3. Median filtering (removes spikes)
smoothed_traces = np.array([medfilt(t, kernel_size=3) for t in filtered_traces])

# Verify improvement
original_corr = np.corrcoef(traces[:,21], intermediates_HW)[0,1]
improved_corr = np.corrcoef(smoothed_traces[:,21], intermediates_HW)[0,1]
print(f"Correlation improved from {original_corr:.3f} to {improved_corr:.3f}")
```

```python
from scipy.signal import correlate

reference = smoothed_traces[0]  # Use filtered trace
aligned_traces = [reference]  # First trace as reference

for trace in smoothed_traces[1:]:
    corr = correlate(reference, trace, mode='same')
    shift = np.argmax(corr) - len(reference)//2
    aligned_trace = np.roll(trace, shift)
    aligned_traces.append(aligned_trace)

aligned_traces = np.array(aligned_traces)
aligned_corr = np.corrcoef(aligned_traces[:,21], intermediates_HW)[0,1]
print(f"After alignment: {aligned_corr:.3f}")
```
