# Load Input CSV

```python
# Load all csv
input_data = [
    pd.read_csv(f"{base_dir}/tv_locality_refresh.csv")
    for base_dir in base_dirs
]
```

# Extract output value (e.g., x12) and compute intermediate model

```python
def HW(x):
    return bin(x).count('1')
    
def HD(x, y):
    return bin(x ^ y).count('1')
    
intermediates_list_hw = [
    (df["c0"].apply(lambda x: int(x, 16)) + df["c1"].apply(lambda x: int(x, 16)))
    .apply(HW)
    .to_numpy()[:100]
    for df in input_data
]

intermediates_list_hd = []
for df in input_data:
    # Convert hex strings to integers
    c0_int = df["c0"].apply(lambda x: int(x, 16))
    c1_int = df["c1"].apply(lambda x: int(x, 16))

    # Sum the two registers
    reg_sum = c0_int + c1_int

    # Shift by 1 to get previous value, compute HD with current value
    reg_sum_shifted = reg_sum.shift(1).fillna(0).astype(int)
    reg_sum = reg_sum.astype(int)

    hd_vals = [HD(curr, prev) for curr, prev in zip(reg_sum, reg_sum_shifted)]

    intermediates_list_hd.append(np.array(hd_vals[:100]))
```

# Add Gaussian Noise

```python
def add_noise(traces, snr_db):
    """
    Add Gaussian noise to traces based on the given SNR in dB.
    SNR (Signal-to-Noise Ratio) = signal_power / noise_power
    """
    signal_power = np.mean(traces ** 2)
    snr_linear = 10 ** (snr_db / 10)
    noise_power = signal_power / snr_linear
    noise = np.random.normal(0, np.sqrt(noise_power), traces.shape)
    return traces + noise

# Add noise (e.g., SNR = 5 dB; adjust for more or less noise)
trace_data_HW = [add_noise(traces, snr_db=5) for traces in trace_data_HW]

trace_data_HD = [add_noise(traces, snr_db=5) for traces in trace_data_HD]
```

# Normalize traces

```python
trace_data_HW = [
    (td - td.mean(axis=0)) / td.std(axis=0) for td in trace_data_HW
]
trace_data_HD = [
    (td - td.mean(axis=0)) / td.std(axis=0) for td in trace_data_HD
]
```

# Compute correlation

```python
correlations_list_HW = []

for traces, intermediates in zip(trace_data_HW, intermediates_list_hw):
    print(f"traces shape hw: {traces.shape}, intermediates shape: {intermediates.shape}")

    traces = (traces - traces.mean(axis=0)) / traces.std(axis=0)

    correlations = np.array([
        np.corrcoef(traces[:, t], intermediates)[0, 1]
        for t in range(traces.shape[1])
    ])

    correlations_list_HW.append(correlations)

correlations_list_HD = []

for traces, intermediates in zip(trace_data_HD, intermediates_list_hd):
    print(f"traces shape hd: {traces.shape}, intermediates shape: {intermediates.shape}")

    traces = (traces - traces.mean(axis=0)) / traces.std(axis=0)

    correlations = np.array([
        np.corrcoef(traces[:, t], intermediates)[0, 1]
        for t in range(traces.shape[1])
    ])

    correlations_list_HD.append(correlations)
```

```python
plt.figure(figsize=(30, 6))

colors = ['blue', 'green', 'red', 'purple']
for i, corr in enumerate(correlations_list_HW):
    plt.plot(corr, label=f'locality_ref v{i}', alpha=0.6, color=colors[i % len(colors)])

plt.xlabel("Time Sample")
plt.ylabel("Correlation")
plt.title("Correlation-Based Leakage Detection for HW")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('figures/correlation_HW_all_versions_locality_ref_v1.pdf', bbox_inches='tight')
plt.show()
```

```python
num_versions = len(correlations_list_HW)
cols = math.ceil(num_versions / 2)  # Try to split into 2 rows
rows = 2 if num_versions > 1 else 1

fig, axes = plt.subplots(rows, cols, figsize=(6 * cols, 5 * rows), sharey=True)

# Flatten the axes array for easy iteration
axes = axes.flatten() if num_versions > 1 else [axes]

colors = ['blue', 'green', 'red', 'purple']

for i, (corr, ax) in enumerate(zip(correlations_list_HW, axes)):
    ax.plot(corr, label=f'locality_ref v{i}', alpha=0.6, color=colors[i % len(colors)])
    ax.set_title(f'locality_ref v{i}', fontsize=14)
    ax.set_xlabel("Time Sample", fontsize=12)
    ax.grid(True)
    ax.legend(fontsize=10)

# Set y-label only on leftmost plots
for i in range(0, len(axes), cols):
    axes[i].set_ylabel("Correlation", fontsize=12)

# Hide any unused subplots
for j in range(num_versions, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.savefig('figures/correlation_HW_all_versions_locality_ref_v2.pdf', bbox_inches='tight')
plt.show()
```
